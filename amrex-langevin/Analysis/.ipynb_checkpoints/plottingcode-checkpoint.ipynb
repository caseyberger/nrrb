```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
%matplotlib inline
import seaborn as sns
from defs_cori import *
```

Define plotting parameters

```python
fig_params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 5),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
pylab.rcParams.update(fig_params)

plot_horizontal = 7.5
plot_vertical = 5.5

re_color = "darkblue"
im_color = "darkred"
```

```python
curr_dir = os.getcwd()+'/'
plotdir = curr_dir
```

# Langevin evolution Plotting

import the csv

```python
filename = "nrrb_data_test/logfile_D_2_Nx_41_Nt_160_dt_0.05_nL_1000_eps_0.001_m_1_wtr_0_wz_0_l_0_mu_-0.01.log"
input_filename = "nrrb_data_test/inputs_test"
p = extract_parameters(curr_dir,input_filename)
data_tL = pd.read_csv(curr_dir+filename)
data_tL = data_tL.rename(columns={"#step": "step"})
data_tL = data_tL.drop(columns=["Unnamed: 9"])
```

```python
data_tL.head()
```

Add a column for the Langevin time, steps scaled by the Langevin stepsize

```python
eps = p["eps"]
tL = data_tL.step*eps
data_tL['tL'] = tL
data_tL.head()
```

Plot real and imaginary density as a function of Langevin time to identify thermalization point. Running averages will be slightly off since they are computed without throwing away the thermalization period yet.

```python
fig, ax = plt.subplots()
nRe = data_tL['Re[<n>]']
nIm = data_tL['Im[<n>]']
tL =  data_tL['tL']  
steps = np.arange(1,len(tL)+1)
ra_Re = np.cumsum(nRe)/steps
ra_Im = np.cumsum(nIm)/steps
ax.plot(tL, nRe, color = "darkcyan",alpha =0.5,label='_nolegend_')
ax.plot(tL, nIm, color = "darkmagenta",alpha =0.5,label='_nolegend_')
ax.plot(tL, ra_Re, ls='dashed', color = "darkcyan", label = "Real")
ax.plot(tL, ra_Im, ls='dashed', color = "darkmagenta", label = "Imaginary")
#determine thermalization point and draw a line on the plot to show it
therm_step = 0.05/eps
ax.axvline(x=0.05, color = "darkslategray", label = "$n_{steps} = $"+str(therm_step))
ax.legend()
ax.set_xlabel("$t_{L}$")
ax.set_ylabel(r"$\hat{n}(t_{L})$")
plt.show()
```

# Averaged Observables Plotting

```python
filename = "all_averaged_observables.csv"
data_avg = pd.read_csv(curr_dir+filename)
data_avg.head()
```

### Free Gas Plots

Plot density as a function of mu for different $\epsilon$

```python
Nx = 21
Nt = 160
plot_params = {
    #'mu' : -0.01,
    'Nx' : Nx,
    'Nt' : Nt,
    #'eps' : 0.001,
    #'nL' : 300000.0,
    'wtr' : 0.0, # trap frequecny
    'wz' : 0.0, # rotation frequency
    'lambda' : 0.0, # lambda, contact interaction strength
}
plot_data = data_avg[(data_avg[list(plot_params)] == pd.Series(plot_params)).all(axis=1)]
```

Compute analytical solution

```python
import NRRB_analytical_solutions as nrrb_exact
Nt = 160
Nx = 21
mu_list = np.array(plot_data["mu"].unique())
print(mu_list)
```

```python
#exact solutions
exact_phisq = [nrrb_exact.noninteracting_phisq(2,1.0,mu,Nx,Nt,0.05) for mu in mu_list]
exact_dens = [nrrb_exact.noninteracting_density(2,1.0,mu,Nx,Nt,0.05) for mu in mu_list]
d = {'mu': mu_list, 
     'Re Phisq Exact': [i.real for i in exact_phisq],
     'Im Phisq Exact':[i.imag for i in exact_phisq],
    'Re n Exact':[i.real for i in exact_dens],
     'Im n Exact':[i.imag for i in exact_dens],
    }
free_gas_exact = pd.DataFrame(data = d)
```

```python
fig, ax = plt.subplots(1,2,figsize = (3*plot_horizontal,1.5*plot_vertical))
beta = plot_data['beta'].unique()[0]
x=sorted(mu_list*beta)
exact_phisq_Re = free_gas_exact['Re Phisq Exact']
exact_phisq_Im = free_gas_exact['Im Phisq Exact']
exact_n_Re = free_gas_exact['Re n Exact']
exact_n_Im = free_gas_exact['Im n Exact']
exactReerr = np.zeros_like(exact_n_Re)
exactImerr = np.zeros_like(exact_n_Im)
eps_list = np.array(plot_data["eps"].unique())
ax[0].errorbar(x=x,y=exact_phisq_Re,yerr = exactReerr,
                   label = "Re, exact")#, color = re_color,linestyle='--')
ax[0].errorbar(x=x,y=exact_phisq_Im,yerr = exactImerr,
                   label = "Im, exact")#, color = im_color,linestyle='--')
ax[1].errorbar(x=x,y=exact_n_Re,yerr = exactReerr,
                   label = "Re, exact")#, color = re_color,linestyle='--')
ax[1].errorbar(x=x,y=exact_n_Im,yerr = exactImerr,
                   label = "Im, exact")#, color = im_color,linestyle='--')
for eps in eps_list:
    df = plot_data[plot_data['eps'] == eps]
    df = df.sort_values(by=['mu'])
    nRe = df['Re n']
    nRe_err = df['Re n err']
    nIm = df['Im n']
    nIm_err = df['Im n err']
    phisqRe = df['Re Phisq']
    phisqRe_err = df['Re Phisq err']
    phisqIm = df['Im Phisq']
    phisqIm_err = df['Im Phisq err']
    ax[0].plot(x,phisqRe,label = "Re, $\epsilon =$"+str(eps))#, color = re_color)
    ax[0].fill_between(x,phisqRe-phisqRe_err,phisqRe+phisqRe_err,
                       label="_nolegend_")#, color = re_color, alpha=0.25)
    ax[0].plot(x,phisqIm,label = "Im, $\epsilon =$"+str(eps))#, color = im_color)
    ax[0].fill_between(x,phisqIm-phisqIm_err,phisqIm+phisqIm_err,
                       label="_nolegend_")#, color = im_color,alpha = 0.25)
    ax[0].legend(fontsize=32)
    ax[0].tick_params(labelsize=22)
    ax[0].set_xlabel(r"$\beta \mu$", fontsize=32)
    ax[0].set_ylabel(r"$\langle\phi^{*}\phi\rangle$", fontsize=32)
    ax[1].plot(x,nRe,label = "Re, $\epsilon =$"+str(eps))#, color = re_color)
    ax[1].fill_between(x,nRe-nRe_err,nRe+nRe_err,
                       label="_nolegend_")#, color = re_color, alpha=0.25)
    ax[1].plot(x,nIm,label = "Im, $\epsilon =$"+str(eps))#, color = im_color)
    ax[1].fill_between(x,nIm-nIm_err,nIm+nIm_err,
                       label="_nolegend_")#, color = im_color,alpha = 0.25)
    ax[1].legend(fontsize=32)
    ax[1].tick_params(labelsize=22)
    ax[1].set_xlabel(r"$\beta \mu$", fontsize=32)
    ax[1].set_ylabel(r"$\langle\hat{n}\rangle$", fontsize=32)
plt.tight_layout()
plt.savefig(plotdir+"free_gas_Nx_"+str(Nx)+"_Nt_"+str(Nt)+".png",format ='png')
plt.savefig(plotdir+"free_gas_Nx_"+str(Nx)+"_Nt_"+str(Nt)+".pdf",format ='pdf')
plt.show()
```
