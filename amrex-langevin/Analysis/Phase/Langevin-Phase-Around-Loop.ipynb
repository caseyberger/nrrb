```python
import yt
import numpy as np
```

```python
ds = yt.load("plt00000")
```

```python
def _Theta(field, data):
    # For this debugging test I saved the field phase to the variable phi_2_Re ...
    theta = data[('boxlib', 'phi_2_Re')]
    return theta
ds.add_field("Theta", function=_Theta, take_log=False, units="")
```

```python
slc = yt.SlicePlot(ds, 'z', 'Theta')
#slc.set_zlim("Theta", -3.14, 3.14)
slc.show()
```
