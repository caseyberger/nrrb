```python
import yt
import numpy as np
```

```python
ds = yt.load("plt00000")
```

```python
def _PathDivLoopLength(field, data):
    # For debugging, in this run in the initialization I set phi_1_Re = loop local length / loop total length
    return data[('boxlib', 'phi_1_Re')]

ds.add_field("Path Length/Loop Length", function=_PathDivLoopLength, take_log=False, units="")

def _LoopRadius(field, data):
    # For debugging, in this run in the initialization I set phi_1_Im = loop radius
    return data[('boxlib', 'phi_1_Im')]

ds.add_field("Loop Radius", function=_LoopRadius, take_log=False, units="")
```

```python
slc = yt.SlicePlot(ds, 'z', "Path Length/Loop Length")
slc.set_log("Path Length/Loop Length", False)
slc.set_zlim("Path Length/Loop Length", 0.0, 1.0)
```

```python
slc = yt.SlicePlot(ds, 'z', "Loop Radius")
slc.set_log("Loop Radius", False)
```
